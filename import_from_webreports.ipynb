```python
import pandas as pd
import re

path_csv = "/Users/alexandercappelen/Downloads/WebReports-Export-2025-8-8.csv"

# manual match id you assign
MATCH_ID = "Friheden_Hvidovre_2025_05_13"

# read raw file
with open(path_csv, "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]

sections = []
current_section = None
buffer = []

for line in lines:
    # detect new section
    if re.match(r"^GAME\d+:", line) or line.startswith("MATCH:") or line.startswith("AGGREGATE"):
        # save previous buffer
        if current_section and buffer:
            df = pd.DataFrame([row.split(",") for row in buffer[1:]], columns=buffer[0].split(","))
            df["section"] = current_section
            df["match_id"] = MATCH_ID
            sections.append(df)
        # reset
        current_section = line
        buffer = []
    else:
        buffer.append(line)

# save last section
if current_section and buffer:
    df = pd.DataFrame([row.split(",") for row in buffer[1:]], columns=buffer[0].split(","))
    df["section"] = current_section
    df["match_id"] = MATCH_ID
    sections.append(df)

# combine everything
all_data = pd.concat(sections, ignore_index=True)

# parse section info
def parse_section(s):
    if s.startswith("GAME"):
        game = re.match(r"GAME(\d+): (.+) \((\d+) - (\d+)\)", s)
        return {"set_id": int(game[1]), "match_name": game[2], "score_us": int(game[3]), "score_them": int(game[4])}
    elif s.startswith("MATCH:"):
        return {"set_id": "MATCH", "match_name": s.replace("MATCH: ", ""), "score_us": None, "score_them": None}
    elif s.startswith("AGGREGATE"):
        return {"set_id": "AGGREGATE", "match_name": None, "score_us": None, "score_them": None}
    return {}

meta = all_data["section"].apply(parse_section).apply(pd.Series)
tidy_data = pd.concat([all_data, meta], axis=1).drop(columns=["section"])

# example tidy result
print(tidy_data.head())
```

```python
import pandas as pd
import re

def parse_volleyball_stats(filepath):
    """
    Parse a volleyball stats export into separate DataFrames per set/match/aggregate.
    Extracts metadata (opponent, date, score) and adds them as columns.
    Venue is ignored.
    
    Parameters
    ----------
    filepath : str
        Path to the raw CSV-like export file.
    
    Returns
    -------
    dict
        Dictionary of DataFrames keyed by "GAME1", "GAME2", ..., "MATCH", "AGGREGATE"
    """
    
    with open(filepath, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f if line.strip()]
    
    dfs = {}
    current_section = None
    buffer = []
    
    for line in lines:
        # detect section markers
        if re.match(r"^GAME\d+:", line) or line.startswith("MATCH:") or line.startswith("AGGREGATE"):
            # save previous section if exists
            if current_section and buffer:
                df = pd.DataFrame([row.split(",") for row in buffer[1:]], columns=buffer[0].split(","))
                dfs[current_section] = df
            # reset
            current_section = line
            buffer = []
        else:
            buffer.append(line)
    
    # save last section
    if current_section and buffer:
        df = pd.DataFrame([row.split(",") for row in buffer[1:]], columns=buffer[0].split(","))
        dfs[current_section] = df
    
    # --- Clean up keys + add metadata ---
    cleaned_dfs = {}
    for key, df in dfs.items():
        if key.startswith("GAME"):
            game_match = re.match(r"GAME(\d+): [^-]+-(.+)-(\d{4}-[A-Za-z]+-\d{2}) \((\d+) - (\d+)\)", key)
            if game_match:
                set_num = int(game_match.group(1))
                opponent = game_match.group(2)
                date = game_match.group(3)
                score_us = int(game_match.group(4))
                score_them = int(game_match.group(5))
                
                df = df.copy()
                df["set_id"] = set_num
                df["opponent"] = opponent
                df["date"] = date
                df["score_us"] = score_us
                df["score_them"] = score_them
                cleaned_dfs[f"GAME{set_num}"] = df
        
        elif key.startswith("MATCH:"):
            # Example: "MATCH: Friheden-Hvidovre-2025-May-13"
            match_match = re.match(r"MATCH: [^-]+-(.+)-(\d{4}-[A-Za-z]+-\d{2})", key)
            if match_match:
                opponent = match_match.group(1)
                date = match_match.group(2)
                
                df = df.copy()
                df["set_id"] = "MATCH"
                df["opponent"] = opponent
                df["date"] = date
                df["score_us"] = None
                df["score_them"] = None
                cleaned_dfs["MATCH"] = df
        
        elif key.startswith("AGGREGATE"):
            df = df.copy()
            df["set_id"] = "AGGREGATE"
            df["opponent"] = None
            df["date"] = None
            df["score_us"] = None
            df["score_them"] = None
            cleaned_dfs["AGGREGATE"] = df
    
    return cleaned_dfs




path_csv = "/Users/alexandercappelen/Downloads/WebReports-Export-2025-8-8.csv"

dfs = parse_volleyball_stats(path_csv)

# Add your own match ID afterwards
for key in dfs:
    dfs[key]["match_id"] = "111111"

# Access individual sets
game1 = dfs.get("GAME1")
game2 = dfs.get("GAME2")
match_total = dfs.get("MATCH")
agg_total = dfs.get("AGGREGATE")

print(game2.head())
```
