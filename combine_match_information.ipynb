```python
import pandas as pd
import numpy as np


results = pd.read_csv("match-results-dvbf.csv")
stats = pd.read_csv("match-statistics.csv")

results["date"] = results["date"].str.strip()
results["date"] = pd.to_datetime(results["date"])
```

```python
results

# From "results", i want:
# - kampnr (key)
# - date
# - opponent
# - played
# - result
# - result_for_team

```

```python
df_stats = stats.drop(columns=["opponent"])

results_played = results[results["resultat"].notna()]

df_stats_with_id = df_stats.merge(
    results_played[["kampnr", "hjemme", "ude", "spillested", "played", "resultat", "result_for_frb", "date", "match_name", "opponent"]],
    on="kampnr",
    how="left"
)

df_stats_with_id
```

```python
df = stats 

team_abbr = {
    "Frederiksberg Volley": "frb",
    "Team Køge": "tkv",
    "Amager Volley.2": "avk",
    "Amager": "avk",
    "Lyngby-Gladsaxe VK": "lgv",
    "Hvidovre VK.2": "hv2",
    "Gentofte Volley.2": "gv2",
    "Farum-Holte": "fah"
    # Add more teams as needed
}

# fallback: first 3 letters lowercase if not in dictionary
def abbrev(team_name):
    return team_abbr.get(team_name, team_name[:3].lower())

# Create match label
df["match_name"] = df.apply(
    lambda row: f"frb-{abbrev(row['opponent'])}" if row["frb_home_away"] == "home"
                else f"{abbrev(row['opponent'])}-frb",
    axis=1
)


results_played = results[results["resultat"].notna()]

df_stats_with_id = df.merge(
    results_played[["kampnr", "hjemme", "ude", "spillested", "played", "resultat", "result_for_frb", "date", "match_name", "opponent"]],
    on="match_name",
    how="left"
)

df_stats_with_id["kampnr"] = df_stats_with_id["kampnr"].astype("Int64")
df_stats_with_id
```
